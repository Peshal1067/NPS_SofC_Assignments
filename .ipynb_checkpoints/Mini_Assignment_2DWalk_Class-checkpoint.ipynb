### Mini Assignment:
Create a 2DWalk class and implement above functions as member function of class. Implement Class to create multiple random walk objects and visualize the 3 walks data

### Solution:

```python
import numpy as np
import random as random
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

#### 1.Creating a 2D Random walk **class**

```python
class R_walk2D(object):
    
    def __init__(self,N,x0,y0):
        self.N = N
        self.X = []
        self.Y = []
        self.x0 = x0
        self.y0 = y0
                
    def move2D(self):
        '''Function that moves a point to one step  in 2D'''
        direction =  random.choice(['x','y'])
        if direction == 'x':
            r = random.choice([-1,1])
            if r ==1:x = self.x0+1
            else: x = self.x0 -1
            y = self.y0
        if direction == 'y':
            r = random.choice([-1,1])
            if r ==1:y = self.y0+1
            else: y = self.y0 -1
            x = self.x0
        return x,y
    
    def walk2D(self):
        '''Function that makes a point walk N step in 2D'''
           
        k=0
        while k < self.N:
            x,y = self.move2D()                                         # Move a step
            self.X.append(x);self.Y.append(y)                           # Collect X and Y coordinates
            self.x0 = x; self.y0 = y                                    # Setting previous position for the next step
            k = k+1
        
        return self.X,self.Y  
      
```

```python
'''Checking worability of the class''' 

X = R_walk2D(N = 1000,x0 = 0, y0 = 0)                                
print(X.walk2D())
```

```python
A = R_walk2D(N = 10000,x0 = 0, y0 = 0).walk2D()
B = R_walk2D(N = 10000,x0 = 0, y0 = 0).walk2D()
C = R_walk2D(N = 10000,x0 = 0, y0 = 0).walk2D()

plt.figure(figsize = [8,8])
plt.plot(A[0],A[1],"-")
plt.plot(B[0],B[1],"-")
plt.plot(C[0],C[1],"-")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.title("Random walk in 2D")
plt.show()
```
