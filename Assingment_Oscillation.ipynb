## Solution

### 1.Python function for random oscillation with amplitude in range(1,5) and frequency in range (0.25,1)

```python
'''implement y = Asin(wt)+Bcos(wt) with random amplitude in (1,5) and freq in (0.25,1)'''

import numpy as np
import matplotlib.pyplot as plt
import random

def oscillator(ts):
    amp = random.choice(np.linspace(1,5,20))
    freq = random.choice(np.linspace(0.25,1,20))
    ys = amp*np.sin(ts*(2*np.pi*freq)) + amp*np.cos(ts*(2*np.pi*freq))
    return ys
```

### 2.Create 20 different oscillations by implementing python function defined above and implement superposition principle to obtain single resultant oscillation.

```python
fs = 100
ts = np.arange(0,60,1/fs)


'''List comprehension: '''

ys = sum([oscillator(ts) for i in range(1,21)])

'''create a plot'''

plt.figure(figsize = [20,6])
plt.plot(ts,ys,label= "Superposed oscillation")
plt.legend()
plt.show()
```
