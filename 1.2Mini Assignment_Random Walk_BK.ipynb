## Mini Assignment
- Define random walk in 2D
- Create a random walk data and save data to file (json/csv)
- Create a plot for random walk in 2D as save the plot as (pdf/png) file

### Solution

```python
''' Importing libraries'''
import matplotlib.pyplot as plt
import seaborn as sns
import random as random
import numpy as np
import pandas as pd
import random as random
%matplotlib inline
sns.set()
```

### A. Random walk data in 2D(*Using Iteration*)

```python
'''if you are in (xi,yi) position go to (xf,yf)'''

'''set up xi,yi'''
xi = 1          # Assigning random position (1,-1) other than (0,0) since(0,o) does not show show random walk
yi = -1         # with given random walk conditions

'''store x and y into X and Y list'''
X = [xi]
Y = [yi]

''' iterate to N steps using for loop'''
for t in range(1000):
        direction =  random.choice(['x','y'])
        if direction == 'x':
            r = random.choice([-1,1])
            xf = X[t-1] + r*xi
            yf = Y[t-1]
        if direction == 'y':
            r = random.choice([-1,1])
            yf = Y[t-1] + r*yi
            xf = X[t-1]
    
        '''remember to update X and Y in every loop'''
        X.append(xf)
        Y.append(yf)
print(X,Y)
```

### A.a. Random walk in 2D with( *use of Function*)

```python
def R_walk2D(N,init_pos):
    
    '''if you are in (xi,yi) position go to (xf,yf)'''

    '''set up xi,yi'''
    xi = init_pos[0]
    yi = init_pos[1]

    '''store x and y into X and Y list'''
    X = [xi]
    Y = [yi]

    ''' iterate to N steps using for loop'''
    for t in range(N):
            direction =  random.choice(['x','y'])
            if direction == 'x':
                r = random.choice([-1,1])
                xf = X[t-1] + r*xi
                yf = Y[t-1]
            if direction == 'y':
                r = random.choice([-1,1])
                yf = Y[t-1] + r*yi
                xf = X[t-1]
    
            '''update X and Y in every loop'''
            X.append(xf)
            Y.append(yf)
    return X,Y
```

### B.a. saving data to json file 

```python
'''Saving data which is output of the direct iterations method i.e. A.a'''

import json as json
R_walk2D_Data = {'x':X , 'y': Y}
with open ('data/R_walk2D_Data.json','w') as f:
    json.dump(R_walk2D_Data,f)
```

```python
''' Saving data which is output of the function R_Walk2D(N,init_pos)i.e. A.b. in .json'''

import json as json
N = 1000
R_walk2D_Data_f = {'x':R_walk2D(N,[1,-1])[0], 'y': R_walk2D(N,[-1,1])[1]}
with open('data/R_walk2D_Data_f.json','w') as f:
    json.dump(R_walk2D_Data_f,f)
```

### B.b. saving data to csv file 

```python
'''Creating Dataframe and saving to .CSV to the data from direct iteration method, A.a'''

R_walk2D_Data = {'x':X , 'y': Y}
R_walk2D_Data = pd.DataFrame(R_walk2D_Data)
R_walk2D_Data.to_csv('data/R_walk2D_Data.csv')

'''Creating Dataframe and saving to .CSV to the data from function method, A.b.'''

R_walk2D_Data_f = {'x':R_walk2D(N,[1,-1])[0], 'y': R_walk2D(N,[-1,1])[1]}
R_walk2D_Data_f = pd.DataFrame(R_walk2D_Data_f)
R_walk2D_Data_f.to_csv('data/R_walk2D_Data_f.csv')

```

### C. Creating and saving the plot as (pdf/png) file

```python
'''Make a plot for X and Y obtained from direct iteration method'''    

plt.figure(figsize = [8,8])
plt.plot(X,Y, color = 'r')
plt.xlabel("X_Walk")
plt.ylabel('Y_Walk')
plt.title("Random walk in 2D",fontsize = 15)
plt.grid(True)
plt.savefig("plot/Random walk in 2D.pdf")
plt.savefig("plot/Random walk in 2D.png")
plt.show()
```

```python
'''Making a plot for X and Y obtained from using a function R_walk2D(N,init_pos)'''

N = 1000
A = R_walk2D(N,[1,-1])

plt.figure(figsize = [8,8])
plt.plot(A[0],A[1],"-")

plt.xlabel("X_Walk")
plt.ylabel('Y_Walk')
plt.title("Random walk in 2D_f",fontsize = 15)
plt.grid(True)
plt.savefig("plot/Random walk in 2D_f.pdf")
plt.savefig("plot/Random walk in 2D_f.png")
plt.show()
```
